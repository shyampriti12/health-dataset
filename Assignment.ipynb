```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
# Load dataset

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("./Health_heart_experimental.csv")
df.head(5)
```

### Line graph

```python

sns.lineplot(data=df, x='age', y='HR', marker='o')
plt.title("Age vs HeartRate")
plt.xlabel("Age")
plt.ylabel("HeartRate")
plt.show()
```

Findings:
The maximum heart-rate generally deacreases with the age.
Besides this, we can also find that there is a lot of fluctuations with the data.

### Bar Plot

```python
plt.figure(figsize=(10, 6))
df.groupby('sex')['weightKg'].mean().plot(kind='bar', color=['blue', 'red'], alpha=0.7)
plt.title("Average weight of Male and Female")
plt.xlabel("Sex (0=Female, 1=Male)")
plt.ylabel("Average weight")
plt.xticks(rotation=0)
plt.show()
```

Findings:
Here, we can see that both the male and female weigh in similar number.

### Pie chart

```python
plt.figure(figsize=(8, 8))
df['sex'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'salmon'], startangle=90)
plt.title("Distribution of Male and Female")
plt.xlabel(" (0=Female, 1=Male)")
plt.show()
```

Finding:
The females are dominating the male in terms of population.

### Box Plot

```python
sns.boxplot(x='sex',y='age',hue='BMI',data=df)
plt.xlabel(" (0=Female, 1=Male)")
```

Finding:
The BMI of the males seems impressive as compared to females.

### Histogram


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='purple', alpha=0.7)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Finding:
Here, the people of age 20 are dominating with their count.

### Scatter Plot

```python
sns.scatterplot(x='age', y='SysBP',hue='sex',style ='indication',data=df)
plt.xlabel(" (0=Female, 1=Male)")
plt.show()
```
